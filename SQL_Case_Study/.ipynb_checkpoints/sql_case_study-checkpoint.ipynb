```python
from sqlalchemy import create_engine
import pandas as pd
engine = create_engine('sqlite:///sqlite_db_pythonsqlite.db')
```

<h3> Questions: </h3>

**Q10:** Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members! 

```python
sql_query = ('SELECT * '
               'FROM '
               '(SELECT facility, SUM(cost) AS revenue '
                  'FROM (SELECT name AS facility, ' 
                             'CASE WHEN memid = 0 THEN guestcost ' 
                             'ELSE membercost ' 
                              'END AS cost '
                        'FROM Bookings '
                            'INNER JOIN Facilities '
                            'USING (facid)) '
                    'AS subquery ' 
                 'GROUP BY facility) '
                 'AS main ' 
              'WHERE revenue < 1000 '
              'ORDER BY revenue;')

df = pd.read_sql_query(sql_query, engine)
df
```

**Q11:** Produce a report of members and who recommended them in alphabetic surname,firstname order

```python
sql_query = ("SELECT ( m1.surname || ', ' || m1.firstname ) AS member, " 
                    "( m2.surname || ', ' || m2.firstname ) AS recommended_by "
               "FROM Members AS m1 "
                    "LEFT JOIN Members AS m2 "
                      "ON m1.recommendedby = m2.memid "
              "ORDER BY member;")

df = pd.read_sql_query(sql_query, engine)
df
```

**Q12:** Find the facilities with their usage by member, but not guests

```python
sql_query = ('SELECT name AS facility, COUNT( bookid ) AS usage_by_members '
               'FROM Bookings '
                    'INNER JOIN Facilities '
                    'USING ( facid ) '
              'WHERE memid !=0 '
              'GROUP BY facility;')
             
df = pd.read_sql_query(sql_query, engine)
df
```

**Q13:** Find the facilities usage by month, but not guests

```python
sql_query = ("SELECT name AS facility, "
                    "strftime('%m', starttime) AS month, "
                    "COUNT(bookid) AS usage_by_members "
               "FROM Bookings "
                    "INNER JOIN Facilities "
                    "USING (facid) "
              "WHERE memid != 0 "
              "GROUP BY facility, month;")
             
df = pd.read_sql_query(sql_query, engine)
df            
```
